```python
import requests
from PIL import Image
from bs4 import BeautifulSoup

url = 'https://kimetsu-no-yaiba.fandom.com/wiki/Characters'
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')
# print(soup.prettify())
charactersRaw = soup.find_all("tbody")
# print(charactersRaw[3].prettify()) # this is harashi

imagesReq = []
links = [ link for link in charactersRaw[3].find_all('a')]
links.pop()
for link in links:
    imgTag = link.find('img')
#     print(imgTag)
    src =  '' if imgTag==None else imgTag.get('data-src')
#     print(src)
#     Image.open(src)
#     if (src != '') :
#         Image.open(requests.get(src, stream=True).raw)
    try:
        imgReq = requests.get(src, stream=True).raw
        imagesReq.append(imgReq)
    except Exception:
        pass

# print(images)
# Image.open(imagesReq[1])
# for img in imagesReq:
#     Image.open(img)
# im1 = 'https://static.wikia.nocookie.net/kimetsu-no-yaiba/images/1/15/Giyu_Anime_Profile.png/revision/latest/scale-to-width-down/130?cb=20191125193924'
# Image.open(requests.get(im1, stream=True).raw)
```
